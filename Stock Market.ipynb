```python
import pandas as pd
import numpy as np
```

```python
apple = pd.read_csv('stocks/AAPL.csv')
amazon = pd.read_csv('stocks/AMZN.csv')
google = pd.read_csv('stocks/GOOGL.csv')
meta = pd.read_csv('stocks/FB.csv')
microsoft = pd.read_csv('stocks/MSFT.csv')
nvidia = pd.read_csv('stocks/NVDA.csv')
```

```python
apple.head()
```

```python
# Combine all the data into a list
dfs = [apple,amazon, google, meta, microsoft, nvidia ]
```

## Create additional columns

```python
# Adding "Moving 50" and "Moving 200" columns to the DataFrames
for df in dfs:
    df['MA50'] = df.Close.rolling(50).mean()
    df['MA200'] = df.Close.rolling(200).mean()

# Adding Previous Day Close column
for df in dfs:
    df["Previous Day Closed"] = df.Close.shift(1)

# Create Change in Price column
for df in dfs:
    df['Change in Price'] = df['Close'] - df['Previous Day Closed'] 

# Create percent change column
for df in dfs:
    df['Percent Change in Price'] = df.Close.pct_change()

# Create previous day volume
for df in dfs:
    df['Previous day Volume'] = df.Volume.shift(1)

# Create  change in volume column
for df in dfs:
    df['Change in Volumn'] = df['Volume'] - df['Previous day Volume']

# Create percent change in volume column
for df in dfs:
    df['Percent Change in Volume'] = df.Volume.pct_change()
```
